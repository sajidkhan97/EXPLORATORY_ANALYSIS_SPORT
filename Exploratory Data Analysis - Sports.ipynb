# Exploratory Data Analysis



## Number of Matches played in each IPL season

```python
plt.figure(figsize=(15,4))
sns.countplot('season',data=df,palette="spring")
plt.title("Number of matches playes in each ipl season")
plt.xlabel("season")
plt.ylabel("matches")
plt.grid()
plt.show()
```

## Numbers of matches won by team

```python
plt.figure(figsize = (15,3))
sns.countplot(x='winner',data=df, palette='hot')
plt.title("Numbers of matches won by team ")
plt.xticks(rotation=80)
plt.xlabel("Teams")
plt.ylabel("No of wins")
plt.show()
```

# Match Result

```python
df['win_by']=np.where(df['win_by_runs']>0,'Bat first','Bowl first')
win=df.win_by.value_counts()
labels=np.array(win.index)
sizes = win.values
plt.figure(figsize=(8,6))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title('Match results')
plt.axis('equal')
plt.show()
```

# Numbers of matches won by batting and bowling first




```python
plt.figure(figsize=(18,8))
sns.countplot('season',hue='win_by',data=df)
plt.title("Number of matches won by the batting and bowling first")
plt.xlabel("season")
plt.ylabel("count")
plt.grid()
plt.show()
```

# Toss result

```python
Toss=df.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
plt.figure(figsize = (10,6))
plt.pie(sizes, labels=labels,autopct='%1.1f%%', startangle=90)
plt.title('Toss result')
plt.axis('equal')
plt.show()
```

```python
# numbers of fours hit by team
four_data=df1[df1['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

# Numbers of fours hit by players

```python
# plot graph on four hit by players
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
plt.title("Numbers of fours hit by players ")
plt.xticks(rotation=50)
plt.xlabel("Player name")
plt.ylabel("No of fours")
plt.show()
```

```python
# we will print no of sixes hit by team
six_data=df1[df1['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

# Numbers of six hit by players

```python
# we will plot graph of six hit by players
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='green')
plt.title("Numbers of six hit by players ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of six",fontsize=15)
plt.show()
```

```python
# print the top 10 leading run scorer in IPL
batsman_score=df1.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```
